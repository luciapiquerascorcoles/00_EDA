### Explicación de las variables: 
        

Posiblemente, no todas las variables vayan a ser importantes a la hora de predecir el precio de una vivienda.

  * `CRIM`: tasa porcentual de crimen per cápita por ciudad.
 
  * `ZN`: parte de tierra presidencial zonificada para lotes de más de 25.000 pies cuadrados (aprox. 2.323m²). Esta variable en muchas ocasiones lo normal es que sea = 0, ya que este tipo de zonas no son comunes.
  * `INDUS`: proporción de acres (4046,8564224 m² o 0,40468564224 hectáreas), de negocios no minoristas por ciudad. Se entiende por negocio no minorista, negocio mayorista, o que requiere una zona extensa de superficie para establecerse, por ejemplo, un centro comercial.
  * `CHAS`: variable 'dummy' del Río Charles (= 1 si el tramo limita con el río; 0 en caso contrario). El río Charles separa la ciudad propiamente dicha de Boston de las ciudades de Cambridge y Watertown, y del barrio de Charlestown, que sí pertenece a Boston. El suelo en las zonas cercanas al río, son mucho más caras. 
  * `NOX`: concentración de óxidos nítricos (por 10 millones). La combustión de carburantes a elevadas temperaturas es la principal fuente artificial de NO.
  * `RM`: número medio de habitaciones por vivienda. (en n
  * `AGE`: proporción (%) de viviendas ocupadas por sus propietarios construidas antes de 1940. Viviendas en edificios antiguos.
  * `DIS`: distancias ponderadas a cinco centros de empleo de Boston.
  * `RAD`: índice de accesibilidad a las carreteras radiales.
  * `TAX`: tasa del impuesto sobre la propiedad de pleno valor por 10.000 dólares.
  * `PTRATIO`: proporción de alumnos-maestro por ciudad.
  * `B`: 1000(Bk - 0.63)^2 donde Bk es la proporción de negros por ciudad. (NO me gusta este indicador).
  * `LSTAT`: % de disminución de la población.
  * `MEDV`: valor medio de las casas ocupadas por sus propietarios en miles de dólares. Esta será la variable target.

### Detección de outliers

#### Histograma de cada una de las variables

```python
real_state.hist(figsize=(25, 25), bins=40, xlabelsize=8, ylabelsize=8, );
# el ; evita la información que sale de matplotlib
```

```python
# Histogramas con densidades:
fig, axs = plt.subplots(ncols=7, nrows=2, figsize=(18, 12))
index = 0
axs = axs.flatten()
for k,v in real_state.items():
    sns.distplot(v, ax=axs[index], color = 'b')
    index += 1
plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=5.0)
```

Observando los histogramas de cada una de las variables, observamos que:

* A partir del 80-85% de ocupación (AGE), aumenta mucho más la frecuencia de viviendas.
* Podemos observar claramente que hay muchos valores atípicos particularmente en las variables `CRIM`, `ZN` y `B` (esta variable yo la eliminaría, me parece bastante racista). 
* Puede decirse que la distribución de la variable de `ZN` y de `CRIM`, están bastante sesgadas justo después de la normalización estándar.
* Las variables `RAD` y `CRIM` están muy sesgadas negativamente. 

Debemos, tras lo observado, hacer un analisis de los outliers que presentan algunas de las variables.
 
Lo haremos mediate gráficos boxplot.

#### Boxplots

Vamos a intentar visualizar los outliers de forma más clara.

```python
# Pintamos todos los gráficos juntos en dos filas, con los títulos 
# de cada variable en el lateral izquierdo.
fig, axs = plt.subplots(ncols=7, nrows=2, figsize=(25, 20))
index = 0
axs = axs.flatten()
for k,v in real_state.items():
    sns.boxplot(y=k, data=real_state, ax=axs[index], palette="Set3")
    index += 1
plt.tight_layout(pad=0.4, w_pad=0.7, h_pad=6.0)
```

La variables `CHAS` recordemos que es binaria, es decir, solo toma valores de 0 y 1, por tanto no se genera boxplot, ni tiene sentido observarla.

Las variables `CRIM`, `ZN`, `RM`, `B`, `LSTAT` y `MEDV`, muestran valores atípicos. Comprobamos el % de los mismos en cada variable.

También podríamos haber representado cada variable por separado con un histograma + densidad y un boxplot a la vez. por ejemplo:

```python
f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw={"height_ratios": (.15, .85)})
sns.boxplot(real_state["MEDV"], ax=ax_box, palette="Set2")
sns.distplot(real_state["MEDV"], ax=ax_hist, color = 'c')
ax_box.set(xlabel='');
```

Box-plot utiliza el método IQR para mostrar datos y valores atípicos, pero para obtener una lista de los valores atípicos identificados, tendremos que utilizar la fórmula matemática y recuperar los datos atípicos:

```python
# Para ver el porcentaje de outliers de cada variable hacemo un for loop
# aplicando la siguiente función:

for k, v in real_state.items():
        q1 = v.quantile(0.25)
        q3 = v.quantile(0.75)
        irq = q3 - q1
        v_col = v[(v <= q1 - 1.5 * irq) | (v >= q3 + 1.5 * irq)]
        perc = np.shape(v_col)[0] * 100.0 / np.shape(real_state)[0]
        print("Column %s outliers = %.2f%%" % (k, perc))
    
```

Vemos que efectivamente las variables que más valores atípicos tienen con `CRIM`, `ZN` y `B`. Por su parte `RM`y `MEDV`,  tienen unn porcentaje medio de outliers, que no afecta en exceso. `LSTAT`, `PTRATIO` y `DIS`tienen un porcentaje bastante bajo.

Los valores atípicos no siempre hay que eliminarlos, habría que saber bien de qué naturaleza son. 

#### Comparación de unas variables con otras

Voy a comparar la variable `MEDV`, es decir, el valor medio de las viviendas, con `RM`, número medio de habitaciones de las viviendas, ya que es su variable con mayor correlación.

```python
x = real_state.RM
y = real_state.MEDV
plt.scatter( x, y, c="greenyellow", alpha=0.2, linewidth=6)
plt.xlabel("nº medio habts. por vivienda")
plt.ylabel("valor medio vivienda")
```

Se observa que la relación entre ambas es lineal, ya que a mayor nº de habitaciones, mayor es el valor de la vivienda, en términos generales. 
Se aprecian valores atípicos.
